```python
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
num_registros = 50
```

```python
np.random.seed(10)
dados = pd.DataFrame({
    'preco_atual': np.random.uniform(10, 100, num_registros),
    'PIB': np.random.uniform(1e12, 2e12, num_registros),
    'preco_futuro': np.random.uniform(10, 100, num_registros)  # Variável alvo
})
```

```python
print(dados.head())
```

```python
x = dados.drop('preco_futuro', axis=1)
y = dados['preco_futuro']
```

```python
arvore_preco_futuro = DecisionTreeRegressor(criterion='squared_error' )
arvore_preco_futuro.fit(x, y)
```

```python
arvore_preco_futuro.feature_importances_
```

```python
from sklearn import tree
previsores = ['preco_atual',  'PIB',  ]
figura, eixos = plt.subplots(nrows = 1,ncols = 1,figsize = (20,20))
tree.plot_tree(arvore_preco_futuro, feature_names= previsores )
```

```python
previsoes = arvore_preco_futuro.predict([[79.418858, 11.867675]])
print(previsoes)
```
